```python
data['Date'] = pd.to_datetime(data['Date']).dt.date
```

```python
data = data.dropna()
```

```python
Traffic_By_Area = data.groupby(by='Area Name')['Traffic Volume'].sum()
```

```python
Traffic_By_Area = Traffic_By_Area.apply(lambda x: f'{x:,.0f}')
print(Traffic_By_Area)
```

```python
plt.figure(figsize=(16, 12))

ax = sns.barplot(x='Area Name', y='Traffic Volume', data=data, hue='Area Name', palette='pastel')

for p in ax.patches:
    ax.annotate(f'{p.get_height():,.0f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                fontsize=10, color='black', 
                xytext=(0, 10), textcoords='offset points')

plt.xticks(rotation=45)
plt.show()
```
